```python
!pip install pandas_datareader
from pandas_datareader import data
import pandas as pd
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')

df = table[0]

tickers = list(df['Symbol'])
dict_mktcap = {}
for i,ticker in enumerate(tickers): #not using comprehension bc may choke on new tickers
    print(i/505) #progress
    print(ticker) #progress
    if ".B" not in ticker and 'MOS' not in ticker:
        dict_mktcap[ticker] = data.get_quote_yahoo(ticker)['marketCap'][0]

#merge market cap and ticker list
df2 = pd.DataFrame.from_dict(dict_mktcap,orient='index').reset_index()
df2.columns = ['Symbol','mkt_cap']
df_merge = pd.merge(df,df2)

#filter double googl
df_merge[df_merge['Symbol']!='GOOGL']
df_merge = df_merge[df_merge['Symbol']!='GOOGL']
df_merge.to_csv('implementation_ticker_list.csv')

#slice top 2 of each sector for testing
df_merge = df_merge.groupby(['GICS Sector','Symbol']).agg({'mkt_cap':sum})
df_merge = df_merge['mkt_cap'].groupby(level=0, group_keys=False)
df_merge.nlargest(2).reset_index().to_csv("test_ticker_list.csv")

!git clone https://gist.github.com/dc7e60aa487430ea704a8cb3f2c5d6a6.git /tmp/colab_util_repo
!mv /tmp/colab_util_repo/colab_util.py colab_util.py 
!rm -r /tmp/colab_util_repo
from colab_util import *
drive_handler = GoogleDriveHandler()

drive_handler.upload('test_ticker_list.csv', parent_path='',overwrite=False)
drive_handler.upload('implementation_ticker_list.csv', parent_path='',overwrite=False)
```
